Next we need to convert the date and time to an integer so we can use them for KNN imputation. The way to do this is use **time since epoch.** (Google is your friend).

In this case our original date time reader failed us. Even though it merged the columns, it did not create a datetime type. This is due in large part to the non-standard time format. Let's correct both of those problems now.

```python
import datetime as dt
from dateutil.parser import parse

def dt2epoch(value):
    d = parse(value)
    epoch = (d - dt.datetime(1970,1,1)).total_seconds()
    return epoch    


def colonify(value):
    vals = value.split(' ')
    vals[1] = vals[1].replace('.', ':')
    return ' '.join(vals)
new_df['Date_Time'] = new_df['Date_Time'].apply(colonify)
new_df['Epoch'] = new_df['Date_Time'].apply(dt2epoch)
```
